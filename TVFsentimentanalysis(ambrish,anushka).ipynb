```python
import string 
string.punctuation
```

```python
def remove_punc(text):
    for char in text:
        if char in string.punctuation:
            text=text.replace(char,'') 
    return text
```

```python
amba['comments']=amba['comments'].apply(remove_punc) 
```

```python
from textblob import TextBlob 
TextBlob(ok).sentiment.polarity 
```

```python
def ran_polarity(amb):
    try:
        return TextBlob(amb).sentiment.polarity
    except:
        return 0 
amba['polarity']=amba['comments'].apply(ran_polarity) 
amba.head()
```

```python
and11=amba['polarity'].value_counts()
and11
```

```python
and12=and11.reset_index()
and12.rename(columns={'index':'0-1(+ve sentiment),0 to -1(-ve sentiment)','polarity':'number_of_records'},inplace=True)
and12
```

```python
and12.set_index('0-1(+ve sentiment),0 to -1(-ve sentiment)',inplace=True)
and12 
```
